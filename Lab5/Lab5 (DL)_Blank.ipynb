# Feature Scaling (Mandatory for SVR)**
SVR is highly sensitive to the range of data points. If we don't scale (normalize) the data, the model will fail to find the correct hyperplane.

```python
import sklearn as s
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalerx=StandardScaler()
scalery=StandardScaler()
x_scale=scalerx.fit_transform(x)
y_scale=scalery.fit_transform(y)
print("scaling done")
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scale,y_scale,test_size=0.2,random_state=42)
```

# Fitting SVR on 3 Different Kernel on dataset

```python
from sklearn.svm import SVR #SVR (Support Vector Regression)
#Ordinary regression → tries to minimize total error
# SVR → draws a tube around the function and says:
# “As long as points stay inside this tube, I’m happy.” This tube is called epsilon (ε).
```

```python
svr=SVR()#object
# Default SVR model
# Kernel = 'rbf' (non-linear)
svrl=SVR(kernel='linear') #Linear SVR ,Fits a straight line
svrp=SVR(kernel='poly') # polynomila svr Fits a curved relationships
```

```python
svr.fit(x_scale,y_scale.ravel()) #ravel -flattens a 2D array into a 1D array. flatten()-Always returns copy
```

```python
svrl.fit(x_scale,y_scale.ravel())
```

```python
svrp.fit(x_scale,y_scale.ravel())
```

```python
y_pred=scalery.inverse_transform(svr.predict(x_scale).reshape(-1,1))
```

```python
y_predl=scalery.inverse_transform(svrl.predict(x_scale).reshape(-1,1))
```

```python
y_predp=scalery.inverse_transform(svrp.predict(x_scale).reshape(-1,1))
```

```python
plt.scatter(x,y)
plt.scatter(x,y_pred)
```

```python
plt.scatter(x, y, color='red', label='Real Data')
plt.plot(x, y_pred, color='blue', label='SVR (RBF)')
plt.title('SVR Regression Results')
plt.xlabel('Year / Feature')
plt.ylabel('Population / Target')
plt.legend()
plt.show()
```

### **Student Activity : Prediction**
**Task:** Predict the value for the Year **2025** (or value 6.5 in the demo data). 
Remember: You must transform the input before predicting, and inverse transform the output.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR

# 1. Define scalers
scaler_x = StandardScaler()
scaler_y = StandardScaler()

# 2. Fit scalers on TRAINING data
x_train_scaled = scaler_x.fit_transform(x_train)
y_train_scaled = scaler_y.fit_transform(
    y_train.reshape(-1,1)
).ravel()

# 3. Train model on SCALED data
svr = SVR(kernel='rbf')
svr.fit(x_train_scaled, y_train_scaled)

# 4. Student Activity prediction
year_2025_scaled = scaler_x.transform([[2025]])
pred_scaled = svr.predict(year_2025_scaled)

predicted_value = scaler_y.inverse_transform(
    pred_scaled.reshape(-1,1)
)

print("The predicted value for input 2025 is:", predicted_value[0][0])
#The value 0.2316 is in standardized scale. Since inverse transformation was applied using an unfitted scaler, the result remained in normalized space.
```
