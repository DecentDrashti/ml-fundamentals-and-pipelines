## My First Neuron

```python
import numpy as np

#output 1 if positive 0 ,if negative.
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0

print("Setup complete.")
```

## Part 1 or gate

```python
#inputs
X_or=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
#targets
y_or= np.array([0,1,1,1])
```

```python
#step2 initialise weights
weights=np.array([0,0])
bias=0
lr=0.5
print(f"Initial Weights: {weights}")
print(f"Initial Bias: {bias}")
```

## step 3 manual calculation

```python
#example input:(0,1)
input_sample = X_or[1]
target_value = y_or[1]

#1.weighted sum(dot product)
z=np.dot(input_sample,weights) + bias
print(f"Math: ({input_sample[0]} * {weights[0]}) + ({input_sample[1]} * {weights[1]}) + {bias} ")
print(f"sum(z): {z:.2f}")

#2.activation
prediction = step_function(z)
print(f"prediction: {prediction}")

#3. error
error = target_value-prediction
print(f"error: {error}")
```

```python
## step 4 training
```

```python
epochs = 5 
for epoch in range(epochs):
    print(f"\nEPOCH: {epoch+1}")
    for i in range(4):
        x = X_or[i]
        y = y_or[i]
    
        #predict
        z=np.dot(x,weights) + bias
        pred = step_function(z)
    
        #error
        error=y-pred
    
        #learn(update weights)
        if error != 0:
            weights =  weights + (lr*error*x)
            bias = bias + (lr*error*1) #always one
            print(f"incorrect output for {x}! expected: {y} output: {pred}")
            print(f"updated weights :{weights} bias:{bias}")
        else:
            print(f"corrected for {x}")
```
